```python
import pandas as pd
import numpy as np
import plotly as py
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
import random
sns.set()
```

```python
bff = pd.read_csv('experiments_results/reentrega/todos_bruteforce.csv', sep=";")
bff.t = bff.t.str[0:-2].astype(float)
bf = bff.groupby(["function","n"]).mean().reset_index()
```

```python
mitmf = pd.read_csv('experiments_results/reentrega/todos_MITM.csv', sep=";")
mitmf.t = mitmf.t.str[0:-2].astype(float)
mitm = mitmf.groupby(["function","n"]).mean().reset_index()
```

```python
f, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10), sharey="col", sharex=True)
bf.groupby('function').plot(x="n", y="t", ax=axes[0][0], title="(a) Fuerza bruta, escala 1:1")
bf.groupby('function').plot(x="n", y="t", ax=axes[0][1], title="(b) Fuerza bruta, escala logarÃ­tmica", logy=True)

mitm.groupby('function').plot(x="n", y="t", ax=axes[1][0], title="(c) MitM, escala 1:1")
mitm.groupby('function').plot(x="n", y="t", ax=axes[1][1], title="(d) MitM, escala logarÃ­tmica", logy=True)
```
