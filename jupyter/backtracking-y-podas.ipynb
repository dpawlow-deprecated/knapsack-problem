```python
import pandas as pd
import numpy as np
import plotly as py
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
import random
sns.set()
```

Importo los datos de la tabla. Saco los "ms" de la columna tiempo:

```python
bf = pd.read_csv('experiments_results/backtracking_1.csv', sep=";")
bf.t = bf.t.str[0:-2].astype(float)
```

Saco el promedio para cada n de cada versión:

```python
podas = bf.groupby(["algoritmo","n"]).mean().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
podas.groupby('algoritmo').plot(x="n", y="t", ax=ax, title="Backtracking y podas")
```

```python
fig, ax = plt.subplots(figsize=(7,7))
podas.groupby('algoritmo').plot(x="n", y="t", ax=ax, title="Backtracking y podas", logy=True, legend=True)
```

```python
f, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
podas.groupby('algoritmo').plot(x="n", y="t", ax=axes[0][0], title="Escala 1:1")
podas.groupby('algoritmo').plot(x="n", y="t", ax=axes[0][1], title="Escala logarítmica", logy=True, legend=True)
```
