```python
import pandas as pd
import numpy as np
import plotly as py
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
import random
sns.set()
```

```python
# w equals n
neqwf = pd.read_csv('experiments_results/reentrega/n_equal_bkp_size.csv', sep=";")
neqwf.t = neqwf.t.str[0:-2].astype(float)
nEqualsW = neqwf.groupby(["algorithm","n"]).mean().reset_index()
```

```python

items_s1v1_f = pd.read_csv('experiments_results/reentrega/items_size1_value1_nEqualsW.csv', sep=";")
items_s1v1_f.t = items_s1v1_f.t.str[0:-2].astype(float)
items_s1v1 = items_s1v1_f.groupby(["algorithm","n"]).mean().reset_index()
```

```python
items_s1_f = pd.read_csv('experiments_results/reentrega/items_size1_nEqualsW.csv', sep=";")
items_s1_f.t = items_s1_f.t.str[0:-2].astype(float)
items_s1 = items_s1_f.groupby(["algorithm","n"]).mean().reset_index()
```

```python
bkp_size_100_f = pd.read_csv('experiments_results/reentrega/bkp_size_100.csv', sep=";")
bkp_size_100_f.t = bkp_size_100_f.t.str[0:-2].astype(float)
bkp_size_100 = bkp_size_100_f.groupby(["algorithm","n"]).mean().reset_index()
```

```python
bkp_size_10000_f = pd.read_csv('experiments_results/reentrega/bkp_size_10000.csv', sep=";")
bkp_size_10000_f.t = bkp_size_10000_f.t.str[0:-2].astype(float)
bkp_size_10000 = bkp_size_10000_f.groupby(["algorithm","n"]).mean().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
nEqualsW.groupby('algorithm').plot(x="n", y="t", ax=ax, title="N equals Bkp Size")
```

```python
f, axes = plt.subplots(nrows=4, ncols=2, figsize=(10,15), sharey='col')
nEqualsW.groupby('algorithm').plot(x="n", y="t", ax=axes[0][0], title="Escala 1:1")
nEqualsW.groupby('algorithm').plot(x="n", y="t", ax=axes[0][1], title="Escala 1:1", logy=True)

items_s1.groupby('algorithm').plot(x="n", y="t", ax=axes[1][0], title="")
items_s1.groupby('algorithm').plot(x="n", y="t", ax=axes[1][1], title="", logy=True, legend=True)

items_s1v1.groupby('algorithm').plot(x="n", y="t", ax=axes[2][0], title="")
items_s1v1.groupby('algorithm').plot(x="n", y="t", ax=axes[2][1], title="", logy=True, legend=True)

bkp_size_10000.groupby('algorithm').plot(x="n", y="t", ax=axes[3][0], title="")
bkp_size_10000.groupby('algorithm').plot(x="n", y="t", ax=axes[3][1], title="", logy=True, legend=True)
```

```python
expBkpf = pd.read_csv('experiments_results/reentrega/exponential_bkp_size.csv', sep=";")
expBkpf.t = expBkpf.t.str[0:-2].astype(float)
expBkp = expBkpf.groupby(["algorithm","n"]).mean().reset_index()
```

```python
f, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5), sharey='col')
expBkp.groupby('algorithm').plot(x="n", y="t", ax=axes[0], title="(a) Escala 1:1")
expBkp.groupby('algorithm').plot(x="n", y="t", ax=axes[1], title="(b) Escala logarítmica", logy=True, legend=True)
```

```python
f, axes = plt.subplots(nrows=1, ncols=4, figsize=(20,5), sharey=True)
nEqualsW.groupby('algorithm').plot(x="n", y="t", ax=axes[0], title="(a) W = n", logy=True)

items_s1.groupby('algorithm').plot(x="n", y="t", ax=axes[1], title="(b) Items tamaño 1", logy=True)

items_s1v1.groupby('algorithm').plot(x="n", y="t", ax=axes[2], title="(c) Items tamaño y valor 1", logy=True)

bkp_size_10000.groupby('algorithm').plot(x="n", y="t", ax=axes[3], title="(d) W = 10000", logy=True)
```
