```python
import pandas as pd
import numpy as np
import plotly as py
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
import random
sns.set()
```

```python
# w equals n
neqwf = pd.read_csv('experiments_results/reentrega/backtracking/bt_nEqualBkpSize.csv', sep=";")
neqwf.t = neqwf.t.str[0:-2].astype(float)
nEqualsW = neqwf.groupby(["algorithm","n"]).mean().reset_index()
```

```python

items_s1v1_f = pd.read_csv('experiments_results/reentrega/ultimo.csv', sep=";")
items_s1v1_f.t = items_s1v1_f.t.str[0:-2].astype(float)
items_s1v1 = items_s1v1_f.groupby(["algorithm","n"]).mean().reset_index()
```

```python
items_s1_f = pd.read_csv('experiments_results/reentrega/backtracking/bt_itemsSize1.csv', sep=";")
items_s1_f.t = items_s1_f.t.str[0:-2].astype(float)
items_s1 = items_s1_f.groupby(["algorithm","n"]).mean().reset_index()
```

```python
bkp_size_10000_f = pd.read_csv('experiments_results/reentrega/backtracking/bt_fixedBkpSize.csv', sep=";")
bkp_size_10000_f.t = bkp_size_10000_f.t.str[0:-2].astype(float)
bkp_size_10000 = bkp_size_10000_f.groupby(["algorithm","n"]).mean().reset_index()
```

```python
f, axes = plt.subplots(nrows=2, ncols=4, figsize=(20,10), sharey='row', sharex=True)
nEqualsW.groupby('algorithm').plot(x="n", y="t", ax=axes[0][0], title="(a) W = n")
nEqualsW.groupby('algorithm').plot(x="n", y="t", ax=axes[1][0], title="(a) Escala logarítmica", logy=True)

items_s1.groupby('algorithm').plot(x="n", y="t", ax=axes[0][1], title="(b) Items tamaño 1")
items_s1.groupby('algorithm').plot(x="n", y="t", ax=axes[1][1], title="(b) Escala logarítmica", logy=True, legend=True)

items_s1v1.groupby('algorithm').plot(x="n", y="t", ax=axes[0][2], title="(c) Items tamaño y valor 1")
items_s1v1.groupby('algorithm').plot(x="n", y="t", ax=axes[1][2], title="(c) Escala logarítmica", logy=True, legend=True)

bkp_size_10000.groupby('algorithm').plot(x="n", y="t", ax=axes[0][3], title="(d) W = 10000")
bkp_size_10000.groupby('algorithm').plot(x="n", y="t", ax=axes[1][3], title="(d) Escala logarítmica", logy=True, legend=True)
```
